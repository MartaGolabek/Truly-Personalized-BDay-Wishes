```python
import pandas as pd
import numpy as np
import os
```

```python
data_set_paths = os.listdir("../data/raw/")
len(data_set_paths)
```

```python
raw_data = []
for path in data_set_paths:
    data_set = pd.read_csv("../data/raw/" + path, delimiter=";")
    raw_data.append(data_set)
```

## Determine an average length of bday wishes

```python
def determine_avg_length(data_set):
    avg_len = 0
    for text in data_set.text:
        avg_len = avg_len + len(text)
    return avg_len / len(data_set)
```

```python
i = 1
for data_set in raw_data:
    avg_length = round(determine_avg_length(data_set))
    print(f"Average length of bday wishes in data set {i} amounts to about {avg_length}")
    i = i + 1
```

## Merge all data sets vertically

```python
merged_raw_data = raw_data[0].append(raw_data[1], ignore_index=True).append(raw_data[2], ignore_index=True).append(raw_data[3], ignore_index=True).append(raw_data[4], ignore_index=True)
print(f"There are {len(merged_raw_data)} bday wishes in the data set.")
```

## Check how many of them starts with the phrase "Happy birthday"

```python
counter = 0
for record in merged_raw_data.iterrows():
    if record[1].values[0].lower().startswith('happy birthday'):
        counter = counter + 1
print(f"{counter} out of {len(merged_raw_data)} bday wishes start with the phrase 'happy birthday'")
```

## -> Happy birthday may be used as a starting phrase

## Merge all rows and save the entire text as a txt file

```python
entire_data_set = ''
for record in merged_raw_data.iterrows():
    entire_data_set = entire_data_set + record[1].values[0] + ' '
```

```python
entire_data_set
```

```python
with open('../data/processed/bday_wishes.txt', 'w+') as file:
    file.write(entire_data_set)
```

## Create a new data set after crawling more data

```python
def prepare_data_set(target_path):
    data_set_paths = os.listdir(target_path)
    raw_data = []
    for path in data_set_paths:
        data_set = pd.read_csv(target_path + path, delimiter=";")
        raw_data.append(data_set)
    for i in range(len(data_set_paths)):
        if i == 0:
            merged_raw_data = raw_data[0]
        else:
            merged_raw_data = merged_raw_data.append(raw_data[1], ignore_index=True)
    print(f"There are {len(merged_raw_data)} bday wishes in the data set.")
    return merged_raw_data
```

```python
new_data_set = prepare_data_set("../data/raw/")
```

```python
new_data_set
```

```python
def df_to_txt(df_data):
    entire_data_set = ''
    for record in df_data.iterrows():
        entire_data_set = entire_data_set + record[1].values[0] + ' '
    return entire_data_set
```

```python
txt_data_set = df_to_txt(new_data_set)
txt_data_set
```

```python
with open('../data/processed/bday_wishes_extended.txt', 'w+') as file:
    file.write(txt_data_set)
```
